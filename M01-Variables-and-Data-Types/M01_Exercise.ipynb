## Part 3 — Type Conversion

```python
# Converting between types
print(int("42"))         # str to int
print(float("3.14"))     # str to float
print(str(100))          # int to str
print(int(9.99))         # float to int — truncates, does not round
```

```python
# TRY
# You have three string variables below.
# Convert them to the correct types so the calculation on the last line works.

units_str     = "21"
fee_str       = "500.50"
discount_str  = "0.10"

# Convert here
units    = int(units_str)
fee      = float(fee_str)
discount = float(discount_str)

total = units * fee * (1 - discount)
print(total)
```

## Part 4 — print() and f-Strings

```python
name  = "Ana Reyes"
age   = 19
gpa   = 1.25

# Basic f-string
print(f"Name: {name}")
print(f"Age: {age}")
print(f"GPA: {gpa:.2f}")

# Expression inside {}
print(f"Birth year (approx): {2025 - age}")
```

```python
# TRY
# Using ONLY f-strings, print the output below exactly.
# Do not hardcode values — use the variables provided.
#
# Expected output:
# ==============================
# Name    : Maria Santos
# Course  : BSIT
# GPA     : 1.75
# ==============================

name   = "Maria Santos"
course = "BSIT"
gpa    = 1.75

# Your code here
print("=" * 30)
print(f"Name    : {name}")
print(f"Course  : {course}")
print(f"GPA     : {gpa}")
print("=" * 30)
```

```python
# EXPLORE
# f-strings support alignment formatting.
# Run this and study the output, then try making your own table below.

print(f"{'Name':<20} {'GPA':>6}")
print(f"{'Ana Reyes':<20} {1.25:>6.2f}")
print(f"{'Jose Santos':<20} {1.50:>6.2f}")
print(f"{'Maria dela Cruz':<20} {1.75:>6.2f}")

# Try adding two more rows using your classmates' names (made up is fine)

print("=" * 30)
print(f"{'Name':<20} {'GPA':>6}")
print(f"{'Marcus Delu':<20} {1.00:>6.2f}")
print(f"{'Joya Brie':<20} {2.40:>6.2f}")
print(f"{'Quezo Beso':<20} {3.00:>6.2f}")
print(f"{'Narna Phony':<20} {1.80:>6.2f}")
```

## Part 5 — input()

```python
# input() always returns a string
name = input("What is your name? ")
print(f"Hello, {name}!")
print(f"Type: {type(name)}")
```

```python
# TRY
# Write a program that:
# 1. Asks for the user's name
# 2. Asks for their age (as an integer)
# 3. Asks for their GPA (as a float)
# 4. Prints a formatted summary using f-strings
#
# Sample output:
# ----------------------------
# Name : Juan dela Cruz
# Age  : 20
# GPA  : 1.75
# ----------------------------

name = input("What is your name? ")
age = input("How old are you?")
gpa = input("What is your GPA?")

print("-" * 30)
print(f"Name    : {name}")
print(f"Age     : {int(age)}")
print(f"GPA     : {float(gpa)}")
print("-" * 30)
```
